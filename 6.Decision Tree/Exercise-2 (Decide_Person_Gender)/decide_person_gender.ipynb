```python
import pandas as pd
df = pd.read_csv('decide_person_gender.csv')
df
```

```python
x = df.iloc[:, :-1]
```

```python
x
```

```python
y = df.iloc[:,3]
y
```

```python
from sklearn import tree
```

```python
classify_ = tree.DecisionTreeClassifier()
```

# Fit by Full Dataset

```python
classify_ = classify_.fit(x,y)
```

```python
prediction_ = classify_.predict([[185,80,45]])
```

```python
prediction_
```

# Show The Tree

```python
x.columns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,8))
tree.plot_tree(classify_,filled=True,rounded=True,feature_names=x.columns)
```
